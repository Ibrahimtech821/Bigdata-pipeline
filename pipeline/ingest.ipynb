# ingest.py Notebook
Loads a dataset path from command-line and saves a copy as `data_raw.csv`.

```python
import sys
import pandas as pd
```

```python
def ingest(input_csv: str, output_csv: str = "data_raw.csv") -> None:
    df = pd.read_csv(input_csv)
    df.to_csv(output_csv, index=False)
    print("Saved a raw copy as data_raw.csv.")
```

```python
def main() -> None:
    if len(sys.argv) < 2:
        print("Usage: python ingest.py <input_csv>")
        sys.exit(1)

    ingest(sys.argv[1])
```

```python
if __name__ == "__main__":
    main()
```
